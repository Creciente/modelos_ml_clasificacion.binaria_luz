# **Preprocesamiento**

```python
from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn import set_config
set_config(display="diagram")
```

```python
df['Occupancy'].value_counts(normalize = True)
```

- Problema de Clasificación en la Determinación de Ocupación 1, o No Ocupación 0.
- Binario.
- Desbalanceado. 79% versus 21%


```python
# Definir target y matriz de características
y = df['Occupancy']
X = df.drop(columns='Occupancy')
```

```python
y[0:5]
```

```python
X.head(3)
```

```python
X.info()
```

Todas las columnas se pueden clasificar como ***numéricas***

```python
# Instanciar selectores de columnas
int_selector = make_column_selector(dtype_include='int')
float_selector = make_column_selector(dtype_include='float')
```

```python
# Instanciar StandardScaler
standard_scaler = StandardScaler()

#Imputación
imputer_int= SimpleImputer(strategy='median')
imputer_float= SimpleImputer(strategy='mean')

# Transformar
col_transformer = make_column_transformer((make_pipeline(imputer_int, standard_scaler), int_selector),
                                          (make_pipeline(imputer_float,standard_scaler), float_selector),
                                           remainder = 'passthrough')

pipeline=make_pipeline(col_transformer)

pipeline
```

```python
#Entrenar
pipeline.fit(X)

X_processed = pipeline.transform(X)
```

```python
X_processed = pd.DataFrame(X_processed, columns=pipeline.get_feature_names_out())
```

```python
X_processed.head(3)
```

```python
# Train,test,split
X_train, X_test, y_train, y_test = train_test_split(X_processed,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=10,
                                                    stratify= y)
```

# **Modelos de Clasificación**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
```

## ***Modelo Regresión Logística***

```python
# Aplicación Modelo Regresión Logística
logreg = LogisticRegression(class_weight='balanced',
                            random_state =10,
                            solver='liblinear')
logreg.fit(X_train, y_train)

train_preds_rl = logreg.predict(X_train)
test_preds_rl= logreg.predict(X_test)

train_score_rl = accuracy_score(y_train, train_preds_rl)
test_score_rl = accuracy_score(y_test, test_preds_rl)

print (f'El accuracy de entrenamiento es {train_score_rl:.3f}.')
print (f'El accuracy de testeo es {test_score_rl:.3f}.')
```

```python
# Reporte
print(classification_report(y_test,test_preds_rl))
```

```python
# Grid Search para Regresión Logística y Reposte con mejores valores.
param_grid = {'C': [0.001, 0.1],
              'penalty': ['l1', 'l2'],
              'solver': ['liblinear']}

grid = GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=2)
grid.fit(X_train, y_train)

print(grid.best_params_)

y_pred = grid.predict(X_test)

print(classification_report(y_test, y_pred))
```
